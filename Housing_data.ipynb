```python
for county in counties_dict:
    try:
        counties_dict[county]['median_2015'] = statistics.median(counties_dict[county]['house_prices_2015'])
        counties_dict[county]['median_2016'] = statistics.median(counties_dict[county]['house_prices_2016'])
        counties_dict[county]['median_2017'] = statistics.median(counties_dict[county]['house_prices_2017'])
        counties_dict[county]['median_2018'] = statistics.median(counties_dict[county]['house_prices_2018'])
        counties_dict[county]['median_2019'] = statistics.median(counties_dict[county]['house_prices_2019'])
        counties_dict[county]['median_2020'] = statistics.median(counties_dict[county]['house_prices_2020'])
    except:
        counties_dict[county]['median_2015'] = 0
```

```python
counties = []
medians_2015 = []
medians_2016 = []
medians_2017 = []
medians_2018 = []
medians_2019 = []
medians_2020 = []

for county in counties_dict:
    counties.append(county)
    medians_2015.append(counties_dict[county]['median_2015'])
    medians_2016.append(counties_dict[county]['median_2016'])
    medians_2017.append(counties_dict[county]['median_2017'])
    medians_2018.append(counties_dict[county]['median_2018'])
    medians_2019.append(counties_dict[county]['median_2019'])
    medians_2020.append(counties_dict[county]['median_2020'])  
    
```
